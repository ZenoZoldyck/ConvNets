```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tf.__version__
```

```python
(train_data, train_label), (test_data, test_label) = tf.keras.datasets.mnist.load_data()
```

```python
train_data[0]
```

```python
train_data = np.array(train_data)
```

```python
data = []
for i in range(60000) :
  a = tf.random.normal([100])
  data.append(a)
```

```python
data_1 = np.array(data)
```

```python
data_1.shape
```

```python
train_data_norm = train_data / 255
```

```python
train_data_norm.shape
```

```python
model_2 = tf.keras.Sequential([
   tf.keras.layers.Dense(512, activation = "sigmoid"),
   tf.keras.layers.Dense(512, activation = "sigmoid"),
   tf.keras.layers.Dense(784, activation = "sigmoid"),
   tf.keras.layers.Reshape((28, 28))                          
])
model_2.compile(loss = tf.keras.losses.mse,
              optimizer = tf.keras.optimizers.Adam(lr = 0.0005),
              metrics = ["accuracy"])
history = model_2.fit(data_1, train_data_norm, epochs = 400)
```

```python
data_2 = []
for i in range(1000) :
  a = tf.random.normal([100])
  data_2.append(a)
data_2 = np.array(data_2)
```

```python
y_pred = model_2.predict(data_2)
```

```python
plt.imshow(y_pred[30])
```
